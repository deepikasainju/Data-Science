```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary Data.csv',usecols = ['Years of Experience','Salary'])
# usecols= aafulai chaine columns haru matra line
# salary years of experince ma depend hune vayeko ley tyo duita matra liyeko
dataset = dataset.dropna()
dataset
```

```python
# dataset[['Years of Experience']].values
```

```python
# dataset['Years of Experience']
# list inside list form ma chaine vayeko ley tala ko tarika ley liyeko
```

```python
# independent variable
X = dataset.iloc[:, 0:-1].values
X

# : = row sabai linxa
# 0:-1 = (column value) years of experince matra choose hunxa
```

```python
# dependent variable
Y = dataset.iloc[:,1:].values
Y
```

train size= min 2/3 (100% ma min 70% huna parxa)
test size= 1/3

```python
# data lai train ra test ma split gareko 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 10)
# test_size=0.1  ==total ko 10% data test ko lagi baki 90% chai train ma
# random_state=shuffle (jati value halda ni farak pardaina)
```

```python
print(X_train)
```

```python
total_len = (len(X))
print("All total length = ",total_len)
print("X_train length = ", len(X_train),len(Y_train))
print("X_test length = ", len(X_test))
```

```python
#Training model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)

# x_train= Years of experience
# y_train= Salary
# pattern check gardai janxa model fit garda
# yeti years of experience ko salary yeti hunxa
```

```python
#testing model
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])
print(exp)

# model banexi test gareko

# # predict the salary
predicted_salary = model.predict(exp)[0][0]
print("Your predicted_salary = ",predicted_salary)

predicted_salary1 = model.predict(exp)
print(predicted_salary1)
# list inside list form ma aauxa yo garda
```
